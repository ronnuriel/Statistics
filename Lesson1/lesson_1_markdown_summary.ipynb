# 🎓 Introduction – Binomial Distribution & Core Formulas
### מבוא – התפלגות בינומית והמשוואות המרכזיות

התפלגות בינומית משמשת לתיאור תהליך שבו מתבצעים **n ניסויים עצמאיים**,  
ובכל אחד יש רק שתי תוצאות אפשריות — *הצלחה* או *כישלון*.

#### 💡 דוגמה:
- כל מטופל שורד (הצלחה) או לא (כישלון).
- ההסתברות להצלחה בכל ניסוי היא **p**.
- הניסוי חוזר **n פעמים** (למשל 40 מטופלים).

---

## 📊 פונקציית המסה (PMF)
ההסתברות שנקבל בדיוק **k הצלחות** מתוך **n ניסויים**:
\[
P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}
\]

כאשר:
- \(\binom{n}{k}\) = מספר הדרכים לבחור k הצלחות מתוך n ניסויים.
- \(p^k\) = ההסתברות שכל ה־k ניסויים יצליחו.
- \((1-p)^{n-k}\) = ההסתברות שכל היתר ייכשלו.

---

## 📈 פונקציית ההתפלגות המצטברת (CDF)
\[
F(x) = P(X \le x)
\]
נותנת את ההסתברות לקבל לכל היותר x הצלחות.

---

## 📏 תוחלת ושונות
- **תוחלת (ממוצע):** \(\mathbb{E}[X] = np\)
- **שונות:** \(\mathrm{Var}(X) = np(1-p)\)
- **סטיית תקן:** \(\sigma = \sqrt{np(1-p)}\)

---

## 🎯 שימושים נפוצים
- מחקרים רפואיים (הצלחת טיפול / שרידות חולים)
- ייצור (מספר מוצרים תקינים)
- ניסויים חוזרים במעבדה או בסימולציה

---

## 🧠 מה נלמד במחברת זו:
1. איך להגדיר מודל בינומי בפייתון.  
2. לחשב הסתברות מדויקת (PMF).  
3. לחשב הסתברות מצטברת (CDF).  
4. להשוות בין שיטות שונות.  
5. להציג גרף של ההתפלגות.  
6. ולבסוף – לשחק עם פרמטרים בעצמך (Playground).  


# 📘 Binomial Clinical Trial — Step by Step (Matplotlib 3.6+ compatible)
### סיכום והרצה שורה־שורה (n=40, p=0.2)

במחברת הזו תוכל/י לעבור שלב־שלב: כל תא **מסביר** מה עושים — ואז תא קוד **מריץ** בדיוק את זה.
אם רוצים, אפשר לשנות את הערכים ולשחק עם הפרמטרים.

**הקשר (Null model):**
- יש **n = 40** מטופלים.
- כל מטופל מוגדר "הצלחה" אם שרד ≥ 5 שנים.
- מניחים תחת מודל האפס (Tx לא טוב יותר) הסתברות **p = 0.2** להצלחה למטופל, עצמאית בין מטופלים.
- לכן: \(X \sim \text{Binomial}(n=40, p=0.2)\).

## שלב 1 — מגדירים את הפרמטרים
כאן מגדירים את מספר הניסויים (**n**) ואת ההסתברות להצלחה בכל ניסוי (**p**).

```python
n = 40
p = 0.2
print(f"Parameters set: n={n}, p={p}")
```

## שלב 2 — טוענים כלים לחישובים
נשתמש ב-`scipy.stats.binom` לחישובי PMF/CDF. אם SciPy לא מותקן, נשתמש בפונקציות עזר ידניות.

```python
from math import comb

try:
    from scipy.stats import binom
    have_scipy = True
except Exception as e:
    have_scipy = False
    print("SciPy not available — using manual helpers.")

def pmf_binom_manual(k, n, p):
    return comb(n, k) * (p**k) * ((1-p)**(n-k))

def cdf_binom_manual(k, n, p):
    # P(X <= k) = sum_{i=0..k} PMF(i)
    return sum(pmf_binom_manual(i, n, p) for i in range(0, k+1))

print("Helpers ready. have_scipy =", have_scipy)
```

## שלב 3 — הסתברות מדויקת: בדיוק 16 הצלחות
נחשב \(P(X=16)\).

```python
k = 16
if have_scipy:
    rv = binom(n=n, p=p)
    p_eq_16 = rv.pmf(k)
else:
    p_eq_16 = pmf_binom_manual(k, n, p)

print(f"P(X={k}) = {p_eq_16:.6f}")
```

## שלב 4 — הסתברות מצטברת מזנב ימין: 16 או יותר
נחשב \(P(X \ge 16)\) ע״י סכימת ה-PMF מ-16 עד 40.

```python
if have_scipy:
    p_ge_16_sum = sum(binom.pmf(i, n, p) for i in range(16, n+1))
else:
    p_ge_16_sum = sum(pmf_binom_manual(i, n, p) for i in range(16, n+1))

print(f"P(X >= 16) by summing PMF = {p_ge_16_sum:.6f}")
```

## שלב 5 — אותה הסתברות בשיטה אחרת (CDF)
כאן נשתמש בזהות: \(P(X \ge 16) = 1 - P(X \le 15)\).

```python
if have_scipy:
    p_ge_16_cdf = 1 - rv.cdf(15)
else:
    p_ge_16_cdf = 1 - cdf_binom_manual(15, n, p)

print(f"P(X >= 16) by 1 - CDF(15) = {p_ge_16_cdf:.6f}")

# בדיקת עקביות
assert abs(p_ge_16_sum - p_ge_16_cdf) < 1e-10
print("✅ שתי השיטות נותנות את אותו ערך.")
```

## שלב 6 — גרף PMF לכל הערכים (ללא התראות Deprecation)
נציג את \(P(X=k)\) לכל \(k=0,1,...,40\) באמצעות bar chart תואם Matplotlib 3.6+.

```python
import numpy as np
import matplotlib.pyplot as plt

ks = np.arange(0, n+1)
if have_scipy:
    pmf_vals = binom.pmf(ks, n, p)
else:
    pmf_vals = np.array([pmf_binom_manual(i, n, p) for i in ks], dtype=float)

plt.figure(figsize=(8,4))
plt.bar(ks, pmf_vals)  # no deprecated args
plt.title(f"Binomial PMF — n={n}, p={p}")
plt.xlabel("k (number of successes)")
plt.ylabel("P(X = k)")
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## שלב 7 — Playground: לשחק עם n, p, k
שנה כאן את הערכים ותריץ שוב כדי לבדוק תרחישים אחרים (הגרף יתעדכן).

```python
# ▼ edit these
n_play = 10
p_play = 0.3
k_play = 3

import numpy as np
import matplotlib.pyplot as plt

if have_scipy:
    rv_play = binom(n=n_play, p=p_play)
    p_eq_k_play = rv_play.pmf(k_play)
    p_ge_k_play = 1 - rv_play.cdf(k_play-1)
    pmf_vals_play = rv_play.pmf(np.arange(0, n_play+1))
else:
    p_eq_k_play = pmf_binom_manual(k_play, n_play, p_play)
    p_ge_k_play = 1 - cdf_binom_manual(k_play-1, n_play, p_play)
    pmf_vals_play = np.array([pmf_binom_manual(i, n_play, p_play) for i in range(0, n_play+1)], dtype=float)

print(f"n={n_play}, p={p_play}, k={k_play}")
print(f"P(X={k_play}) = {p_eq_k_play:.6f}")
print(f"P(X >= {k_play}) = {p_ge_k_play:.6f}")

plt.figure(figsize=(8,4))
plt.bar(np.arange(0, n_play+1), pmf_vals_play)
plt.title(f"Binomial PMF — n={n_play}, p={p_play}")
plt.xlabel("k (number of successes)")
plt.ylabel("P(X = k)")
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
